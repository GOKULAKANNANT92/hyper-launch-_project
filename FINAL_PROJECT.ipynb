````python
Q1=df['Final_Score'].quantile(0.25)
Q2=df['Final_Score'].quantile(0.50)
Q3=df['Final_Score'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df['Final_Score']<lower_bound) | (df['Final_Score']>upper_bound)]
print("Outliers in Final_Score:")
print("Lower Bound:",lower_bound)
print("Upper Bound:",upper_bound)
print("outliers:")
print(outliers)
outliers_removed=df[(df['Final_Score']>=lower_bound) & (df['Final_Score']<=upper_bound)]
plt.figure(figsize=(8,6))
sns.histplot(outliers_removed['Final_Score'],bins=20,kde=True,color='green')
plt.title("Distribution of Final_Score")
plt.xlabel("Final_Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
````


No outliers were detected in the Final_Score data based on the IQR method, as all values fall within the calculated bounds. The distribution appears normal and free of extreme values.

**MODEL DEVELOPMENT**

````python
x=df_encoded.drop(columns=['Final_Score'])
y=df_encoded['Final_Score']
````

In this ,X variable contains the input features by  dropping the target values .Y variable contains only the target value which "Final_Score"

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
````



### *   *Here train_test_split() is a function from sklearn that splits the dataset into the following*:


1.  **x_train:** It contains 80 % of the input data to train the model.
2.   **x_test:** It contains 20 % of the input data to test the model.
3. **y_train:** It contains 80 % of the target variable.
4.   **y_test:** It contains 20 % of the target variable.



````python
from sklearn.linear_model import LinearRegression
````

````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

model = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('regressor', LinearRegression())
])

model.fit(x_train, y_train)
````

````python
x_train = x_train.fillna(x_train.mean())
y_train = y_train.fillna(y_train.mean())
````

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
````

````python
x_test = x_test.fillna(x_train.mean())
y_test = y_test.fillna(y_train.mean())
````

PREDICT THE TEST DATA

````python
y_pred=model.predict(x_test)
````

**RESULT PRESENTATION**

````python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
import numpy as np

mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)


print("Mean Absolute Error:",mae)
print("Mean Squared Error:",mse)
print("Root Mean Squared Error:",rmse)
print("R-squared:",r2)
````

````python
plt.scatter(y_test,y_test,color='red')
plt.scatter(y_test,y_pred,color='blue')
plt.plot([y.min(),y.max()],[y.min(),y.max()],'--b',label='perfect prediction line')
plt.xlabel("Actual final_score")
plt.ylabel("predicted final_score")
plt.title("Actual vs Predicted Final_score")
plt.show()
````

The scatter plot demonstrates the relationship between actual and predicted final scores, with the red points representing the true values and the blue points indicating the predicted values, highlighting how closely the predictions align with the perfect prediction line.

````python
residuals = y_test - y_pred

plt.scatter(y_test, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Actual Final_score")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
````

The residual plot show how well the model predicte the actual values ,with the red line indicating where residuals are zero.ideally,residuals should be randomly scattered around this line,indicating a good model fit.
